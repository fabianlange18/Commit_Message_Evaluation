#### Load data, take subset and generate vectors for each message

```python
import pandas as pd

data = pd.read_csv("results.csv")
```

```python
subset_size = 100000

messages = data["message"][:subset_size].tolist()
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")
```

```python
vectors = []

docs = nlp.pipe(messages)

for doc in docs:
    vectors.append(doc.vector)
```

### Try clustering the subset with k-means

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(100)
kmeans.fit(vectors)
kmeans_prediction = kmeans.predict(vectors)

for label in range(kmeans.n_clusters):
    print("\n--------------- Class " + str(label) + " ---------------\n")
    counter = 1
    for i, message in enumerate(messages):
        if counter == 11:
            break
        if kmeans_prediction[i] == label:
            print(str(counter) + ") " + messages[i] + "\n")
            counter += 1
```

A kmeans model was trained on the vectors spacy calculates for each document.

This clustering approach with seems not to be working as a method to distinguish between good and bad commit messages.  

The fact that the samples are not labeled and therefore cannot be supervised making a classification harder in general, it is not even possible to understand the differences between messages from different output classes.
